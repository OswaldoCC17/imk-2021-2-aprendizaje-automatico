## Example with $h(\theta) = \theta_1 X$

```python
X = np.linspace(0,5,6)
y = X
```

```python
h = X
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('X', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.plot(X, h, "b-", linewidth=2)
plt.plot(X, y, "ro", markersize=6)
plt.grid(True)
plt.show()
```

```python
theta1 = 0.5
h = theta1 * X
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('X', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.plot(X, h, "b-", linewidth=2)
plt.plot(X, y, "ro", markersize=6)
plt.grid(True)
plt.show()
```
